```python
import anndata as ad
from rich import print as rprint
from abcex.utils.config import load_config

paths = load_config()
rprint(paths)
paths['data'] = str(paths['data'])+"/"
```

```python
z = {}
z[1] = ad.read_h5ad(paths['data']+'expression_matrices/Zhuang-ABCA-1/20230830/Zhuang-ABCA-1-log2.h5ad', backed='r')
z[2] = ad.read_h5ad(paths['data']+'expression_matrices/Zhuang-ABCA-2/20230830/Zhuang-ABCA-2-log2.h5ad', backed='r')
z[3] = ad.read_h5ad(paths['data']+'expression_matrices/Zhuang-ABCA-3/20230830/Zhuang-ABCA-3-log2.h5ad', backed='r')
z[4] = ad.read_h5ad(paths['data']+'expression_matrices/Zhuang-ABCA-4/20230830/Zhuang-ABCA-4-log2.h5ad', backed='r')

m = {}
m[1] = ad.read_h5ad(paths['data']+'expression_matrices/MERFISH-C57BL6J-638850/20230830/C57BL6J-638850-log2.h5ad', backed='r')
```

```python
for key in [1,2,3,4]:
    print('Number of genes:', z[key].var.size)
```

```python
ligands = ['Npy',  'Sst',  'Vip', 'Tac2', 'Cck', 'Penk', 
           'Crh', 'Cort', 'Tac1', 'Pdyn', 'Pthlh',
           'Pnoc', 'Trh', 'Grp', 'Rln1', 'Adcyap1', 'Nts', 'Nmb']

receptors = ['Npy1r', 'Npy2r', 'Npy5',  'Sstr1', 'Sstr2', 'Sstr3', 'Sstr4',
             'Vipr1', 'Vipr2', 'Tacr3', 'Cckbr', 'Oprd1', 'Oprm1', 'Crhr1', 
             'Crhr2', 'Sstr1', 'Sstr2', 'Sstr3', 'Sstr4', 'Tacr1', 'Oprd1', 
             'Oprk1', 'Oprm1', 'Pth1r', 'Oprl1', 'Trhr', 'Trhr2', 'Grpr',
             'Rxfp1', 'Rxfp2', 'Rxfp3', 'Adcyap1r1', 'Vipr1', 'Vipr2', 
             'Ntsr1', 'Ntsr2', 'Nmbr']
```

```python
# reporting number of ligands and receptors from Smith et al. 2019 available in the ABC spatial

found_ligands = [l for l in ligands if l in z[1].var['gene_symbol'].tolist()]
print(f'{len(found_ligands)} out of {len(ligands)} ligands found in Zhuang data')
print(sorted(found_ligands))
print('----\n')

found_receptors = [r for r in receptors if r in z[1].var['gene_symbol'].tolist()]
print(f'{len(found_receptors)} out of {len(receptors)} receptors found Zhuang data')
print(sorted(found_receptors))
print('----\n')

found_ligands = [l for l in ligands if l in m[1].var['gene_symbol'].tolist()]
print(f'{len(found_ligands)} out of {len(ligands)} ligands found in MERFISH data')
print(sorted(found_ligands))
print('----\n')

found_receptors = [r for r in receptors if r in m[1].var['gene_symbol'].tolist()]
print(f'{len(found_receptors)} out of {len(receptors)} receptors found in MERFISH data')
print(sorted(found_receptors))
print('----')
```
